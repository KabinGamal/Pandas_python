```python
car_sales=pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"]=pd.to_numeric(
      car_sales["Price"].replace(r"[$,.]","",regex=True)
)
```

```python
car_sales["Price"]=car_sales["Price"]/10
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].astype(int)
```

```python
car_sales
```

```python
car_sales["Total Sales"]=car_sales["Price"].cumsum()
```

```python
car_sales
```

```python
car_sales["Date of sale"]=pd.date_range("1/1/2022",periods=len(car_sales))
```

```python
car_sales
```

```python
car_sales.plot(x="Date of sale",
               y="Total Sales",
               ylabel="Price")
```

```python
car_sales.plot(x="Date of sale",
               y="Total Sales",
               kind="scatter")
```

```python
import matplotlib.pyplot as plt

plt.scatter(car_sales["Odometer (KM)"], car_sales["Price"])
plt.xlabel("Odometer (KM)")
plt.ylabel("Price")
plt.title("Odometer vs Price")
plt.show()
```

```python
import numpy as np

x = car_sales["Odometer (KM)"]
y = car_sales["Price"]

# scatter
plt.scatter(x, y)

# trend line
z = np.polyfit(x, y, 1)
# trend line
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x))

plt.xlabel("Odometer (KM)")
plt.ylabel("Price")
plt.title("Trend of Price vs Odometer")
plt.show()
```
